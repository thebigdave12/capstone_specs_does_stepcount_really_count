```python
id_outliers = fitbit_mean_df.groupby('Id')[["TotalSteps", "TotalDistance", "Calories", 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', "SedentaryMinutes"]].sum()
```

```python
id_outliers.head()
```

```python
id_outliers = id_outliers.reset_index(name=[["TotalSteps", "TotalDistance", "Calories", 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', "SedentaryMinutes"]])
```

```python
id_outliers_sorted = id_outliers.sort_values(by='TotalDistance', ascending=True)
```

```python
id_outliers_sorted = id_outliers_sorted.reset_index().rename(columns={'Id': 'Id_Label'})
```

```python
id_outliers_sorted
```

```python
figure = plt.figure(figsize=(12,20))

ax = sns.barplot(data=id_outliers_sorted, y='TotalSteps', x='index', order=id_outliers_sorted['index'])

plt.ylabel('Total Steps Per Person')
plt.xlabel('Participant Id Label')
plt.title('Distribution of total distance by day')
```
